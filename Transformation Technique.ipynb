--When the features are not normally distubred then the we should transform it into Gaussian Transformation

--Some machine learning algorithm assumes that the data is normally distributed(Acuracy and performance)

--Whenever we have skewed distrubution we should or we can use lograthmatic transformation

--If we want to check whether the feature is gaussian or normal distribution then we can use qq plot

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

logarthmic transformation

```python
plot_data(df,'Age')
```

```python
df['Age_log']=np.log(df['Age'])
```

```python
plot_data(df,'Age_log')
```

Here for Age we donot need to transform because as we can see in the above qq plot its in already normalized

```python
plot_data(df,'Fare')
```

```python
df['fare_log']=np.log(df['Fare']+1)
```

```python
plot_data(df,'fare_log')
```

Square Root Transformation

```python
df['Age_squared']=df.Age**2
```

```python
plot_data(df,'Age_squared')
```

Exponential Transformation

```python
df['Age_Transformation']=df.Age**(1/1.2)
plot_data(df,'Age_Transformation')
```

BoxCoc Transformation

t(y) = (y exp(alpha)-1) / alpha

y = response variable,
alpha = Transformation Parameter,
alpha ranges from -5 to 5

```python
from scipy import stats
```

```python
df['Age_Boxcoc'],parameters=stats.boxcox(df['Age'])
```

```python
print(parameters)
```

```python
plot_data(df,'Age_Boxcoc')
```

boxcox give us the good result so as exponential
