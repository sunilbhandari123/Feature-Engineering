### Handeling categorical Missing values

##  Frequency category imputation

If there is less missing values in the column than we can replace it by the most frequent repeated value and because of that the relation between them will not be distorted.


```python
import pandas as pd
import numpy as np
```

```python
raw_data=pd.read_csv('house_pricing.csv',usecols=['Alley','MiscFeature','Fence','PoolQC','GarageCond','GarageQual','GarageType','GarageFinish','FireplaceQu','Electrical','BsmtFinType2','BsmtFinType1','BsmtExposure','BsmtQual','BsmtCond'])
raw_data.head()
```

```python
df=raw_data
```

```python
df.isnull().mean()
```

Here the missing value percentage of bsmqual,bsmtcond,bsmtexposure,bsmtfintype1,electrical,garagetype
garagefinish,garagequal,garagecond have compratively lower missing value we will look by replacing it by the most frequent /repeted one and we will check whether it will distort the variance or not.

```python
df['BsmtQual'].value_counts()
```

```python
df['BsmtQual'].value_counts().index[0]
```

```python
def functionn (df,variable):
    most_frequent_var=df[variable].value_counts().index[0]
    df[variable].fillna(most_frequent_var,inplace=True)
```

```python
for feature in ['GarageCond','GarageQual','GarageType',
                  'GarageFinish','Electrical','BsmtFinType2','BsmtFinType1',
                  'BsmtExposure','BsmtQual','BsmtCond']:
    
    functionn(df,feature)
```

```python
df.isnull().mean()
```

Upto here we have handle the missing cateogorical features with less missing values by using frequency category Imputations Now we will see the categorical feature having more number of missing values

In other features we have many missing values so that we will just replace nan values with missing string

```python
def functioon(df,variable):
    df[variable+'_missing']=np.where(df[variable].isnull(),'Missing',df[variable])
```

```python
for features in ['Alley','FireplaceQu','PoolQC','Fence','MiscFeature']:
    functioon(df,features)
```
